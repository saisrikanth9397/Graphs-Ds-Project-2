```python
import networkx as nx
import pandas as pd
import os
import numpy as np
```

```python
data = '/content/drive/MyDrive/Ds Data'
firstFile = '/content/drive/MyDrive/Ds Data/tx_transfers_10-11-2020_etherum_tx_csv-000000000025.csv'
```

```python
firstFileDF = pd.read_csv(firstFile, usecols= ['from_addr','to_addr','tx_hash','gas_price'])
for file in os.listdir(data):
  with open(os.path.join(data, file)) as f:
    #print(f.name)
    if f.name != firstFile:
      #print('sfs')
      graph1 = pd.read_csv(f, usecols= ['from_addr','to_addr','tx_hash','gas_price'])
      firstFileDF = firstFileDF.append(graph1)
firstFileDF = firstFileDF.reset_index()
```

```python
firstFileDF = firstFileDF.rename(columns={'from_addr': 'source', 'to_addr': 'target'})
```

```python
DG = nx.from_pandas_edgelist(firstFileDF, edge_attr=True ,create_using=nx.DiGraph())
```

```python
dfsDG = nx.dfs_tree(DG)
list(dfsDG)
```

```python
#del DG
```

```python
topoDG = nx.topological_sort(dfsDG)
list(topoDG)
```

```python
longDFSpath = nx.dag_longest_path(dfsDG)
```

```python
len(longDFSpath)
```

```python
longDG = nx.DiGraph(nx.dfs_edges(dfsDG, source=longDFSpath[0]))
```

```python
list(longDG.edges())
```

```python
firstOp = []
for i in list(longDG.edges()):
  row = (i[0], i[1], DG[i[0]][i[1]]['tx_hash'], DG[i[0]][i[1]]['gas_price'])
  firstOp.append(row)
```
